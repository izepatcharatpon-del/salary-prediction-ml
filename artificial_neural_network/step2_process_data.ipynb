```python
import pandas
import seaborn
import numpy
import sklearn.preprocessing
import sklearn.feature_selection
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
data_train = pandas.read_csv('../dataset/high_salary.train.csv').set_index('id')
data_train
```

```python
data_test = pandas.read_csv('../dataset/high_salary.test.csv').set_index('id')
data_test
```

```python
data_live = pandas.read_csv('../dataset/high_salary.live.csv').set_index('id')
data_live
```

```python
data_train.dtypes
```

```python
categorical_features = [
    'workclass', 
    'education', 
    'marital-status', 
    'occupation',
    'relationship', 
    'race', 
    'sex',
    'native-country-code', 
    'native-country', 
]

for feature in categorical_features:
    possible_values = data_train[feature].unique()
    print(f"{feature:20s} {possible_values.size:2d} {possible_values}")
```

```python
data_train['label'].value_counts()
```

```python
data_test['label'].value_counts()
```

```python
data_train.isna().sum(axis='index')
```

```python
data_test.isna().sum(axis='index')
```

```python
data_train = data_train.dropna()
```

```python
data_train.isna().sum(axis='index')
```

```python
data_test = data_test.dropna()
```

```python
data_test.isna().sum(axis='index')
```

```python
categorical_cols = data_live.select_dtypes(include='object').columns
numerical_cols = data_live.select_dtypes(include='number').columns
```

```python
imputer_num = SimpleImputer(strategy='median')
data_live[numerical_cols] = imputer_num.fit_transform(data_live[numerical_cols])
```

```python
imputer_cat = SimpleImputer(strategy='most_frequent')
data_live[categorical_cols] = imputer_cat.fit_transform(data_live[categorical_cols])
```

```python
ordinal_features = ["education", "age-group", "hoursperweek", "education-num"]
```

```python
ordinal_mapping = [sorted(data_train[feature].dropna().unique()) for feature in ordinal_features]
```

```python
print("")
for f, m in zip(ordinal_features, ordinal_mapping):
    print(f"{f}: {m}")
```

```python
ordinal_encoder = OrdinalEncoder(categories=ordinal_mapping)
```

```python
data_train_ordinal = pandas.DataFrame(
    ordinal_encoder.fit_transform(data_train[ordinal_features]),
    columns=ordinal_features,
    index=data_train.index
)
```

```python
data_test_ordinal = pandas.DataFrame(
    ordinal_encoder.transform(data_test[ordinal_features]),
    columns=ordinal_features,
    index=data_test.index
)
data_test_ordinal
```

```python
data_live_ordinal = pandas.DataFrame(
    ordinal_encoder.transform(data_live[ordinal_features]),
    columns=ordinal_features,
    index=data_live.index
)
data_live_ordinal
```

```python
nominal_features = [
    'workclass',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'sex',
    'native-country-code',
    'native-country',
]
```

```python
onehot_encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
onehot_encoder.fit(data_train[nominal_features])
```

```python
data_train_nominal = pandas.DataFrame(
    onehot_encoder.transform(data_train[nominal_features]),
    columns=onehot_encoder.get_feature_names_out(nominal_features),
    index=data_train.index
)
```

```python
data_test_nominal = pandas.DataFrame(
    onehot_encoder.transform(data_test[nominal_features]),
    columns=onehot_encoder.get_feature_names_out(nominal_features),
    index=data_test.index
)
```

```python
data_live_nominal = pandas.DataFrame(
    onehot_encoder.transform(data_live[nominal_features]),
    columns=onehot_encoder.get_feature_names_out(nominal_features),
    index=data_live.index
)
```

```python
numerical_features = [
    'fnlwgt',
    'capitalgain',
    'capitalloss',
]
```

```python
scaler = StandardScaler()
scaler.fit(data_train[numerical_features])
```

```python
data_train_numerical_features = pandas.DataFrame(
    scaler.transform(data_train[numerical_features]),
    columns=numerical_features,
    index=data_train.index
)
data_train_numerical_features
```

```python
data_test_numerical_features = pandas.DataFrame(
    scaler.transform(data_test[numerical_features]),
    columns=numerical_features,
    index=data_test.index
)
data_test_numerical_features
```

```python
data_live_numerical_features = pandas.DataFrame(
    scaler.transform(data_live[numerical_features]),
    columns=numerical_features,
    index=data_live.index
)
data_live_numerical_features
```

```python
label_encoder = LabelEncoder()
data_train_label = pandas.DataFrame(
    label_encoder.fit_transform(data_train["label"]),
    columns=["label"],
    index=data_train.index
)
```

```python
data_test_label = pandas.DataFrame(
    label_encoder.transform(data_test["label"]),
    columns=["label"],
    index=data_test.index
)
```

```python
data_train_final = data_train_numerical_features.join(data_train_ordinal).join(data_train_nominal).join(data_train_label)
data_test_final  = data_test_numerical_features.join(data_test_ordinal).join(data_test_nominal).join(data_test_label)
data_live_final  = data_live_numerical_features.join(data_live_ordinal).join(data_live_nominal)
```

```python
data_train_final.to_csv("./data/features.train.csv")
data_test_final.to_csv("./data/features.test.csv")
data_live_final.to_csv("./data/features.live.csv")
```
