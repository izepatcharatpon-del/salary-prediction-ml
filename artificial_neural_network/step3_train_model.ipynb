```python
import joblib
import json
import os
import pandas as pd
from sklearn.neural_network import MLPClassifier
```

```python
data = pd.read_csv('data/features.train.csv').set_index('id')
```

```python
X_train = data.drop(columns='label')
y_train = data['label']
```

```python
model = MLPClassifier(
    hidden_layer_sizes = (100,),
    activation = 'tanh',
    solver = 'adam',
    alpha = 3e-4,
    learning_rate_init = 0.0005,
    max_iter = 800,
    random_state = 1,
    verbose = True,
    early_stopping = True,
    validation_fraction=0.1,
    n_iter_no_change=30,
)
```

```python
model.fit(X_train, y_train)
```

```python
joblib.dump(model,'./model/model_ann.joblib')
```

```python
config = model.get_params()
```

```python
with open('./model/config.json', 'w') as f:
    json.dump(config, f, indent=4)
```
