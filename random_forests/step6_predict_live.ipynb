```python
import pandas as pd
import joblib
```

```python
data_live_features = pd.read_csv('./data/features.live.csv').set_index('id')
```

```python
model = joblib.load('./model/model_rf.joblib')
```

```python
predictions = model.predict(data_live_features)
```

```python
submission = pd.DataFrame({
    'id': data_live_features.index,
    'prediction': predictions.astype(float)  
})
```

```python
submission.to_csv('./results/G22_predictions.live.csv', index=False, float_format="%.1f")
```
